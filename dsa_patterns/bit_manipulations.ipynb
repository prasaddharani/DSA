136. Single Number

````python
''' 
Example 1:

Input: nums = [2,2,1]

Output: 1

Example 2:

Input: nums = [4,1,2,1,2]

Output: 4
'''
from typing import List, Optional

class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        res = 0
        for num in nums:
            res = res ^ num 
        return res 

solution = Solution()
solution.singleNumber([4,1,2,1,2])
        
````

191. Number of 1 Bits

Brian Kernighanâ€™s Algorithm

````python

'''
n = 12 (1100)

step-1: 
n & (n - 1)
1100 & 1011 = 1000
count = 1

step-2:
n = 1000
n - 1 = 0111
n = n & (n - 1) = 0000
count = 2
'''
class Solution:
    def hammingWeight(self, n: int) -> int:
        count = 0
        while n:
            n = n & (n - 1)
            count += 1
        return count 
    
solution = Solution()
solution.hammingWeight(12)
````

338. Counting Bits

````python
''' 
Input: n = 5
Output: [0,1,1,2,1,2]
Explanation:
0 --> 0
1 --> 1
2 --> 10
3 --> 11
4 --> 100
5 --> 101
'''

class Solution:
    def countBits(self, n: int) -> List[int]:
        dp = [0] * (n + 1)
        for i in range(1, n + 1):
            dp[i] = dp[i >> 1] + (i & 1)
        return dp 
    
    def countBits(self, n):
        dp = [0] * (n + 1)
        offset = 1
        for i in range(1, n + 1):
            if offset * 2 == i:
                offset = i 
            dp[i] = 1 + dp[i - offset]
        return dp
    
solution = Solution()
solution.countBits(5)
````

190. Reverse Bits

````python
''' 
Input: n = 00000010100101000001111010011100
Output:    964176192 (00111001011110000010100101000000)
Explanation: The input binary string 00000010100101000001111010011100 represents the unsigned integer 43261596, so return 964176192 which its binary representation is 00111001011110000010100101000000.
'''

class Solution:
    def reverseBits(self, n: int) -> int:
        res = 0 
        for i in range(32):
            bit = (n >> i) & 1
            res = res | bit << (31 - i)
        return res 
    
solution = Solution()
solution.reverseBits(43261596)
````

201. Bitwise AND of Numbers Range

````python
''' 

26 â†’ 11010
27 â†’ 11011
28 â†’ 11100
29 â†’ 11101
30 â†’ 11110

If you AND all of these, bits that change at any position will become 0 in the result.

So instead of ANDing all, weâ€™ll find the common prefix in binary using right shifts.

left  = 26 â†’ 11010
right = 30 â†’ 11110
shift = 0

Step-1:
left  = 11010 â†’ 1101 (13)
right = 11110 â†’ 1111 (15)
shift = 1

Step-2:
left  = 1101 â†’ 110 (6)
right = 1111 â†’ 111 (7)
shift = 2

Step-3:
left  = 110 â†’ 11 (3)
right = 111 â†’ 11 (3)
shift = 3

Finally:
Now shift back the common prefix:
left << shift = 3 << 3 = 3 * 8 = 24
Binary: 00011 << 3 â†’ 11000

'''

class Solution:
    def rangeBitwiseAnd(self, left: int, right: int) -> int:
        shift = 0
        while left != right:
            left = left >> 1
            right = right >> 1
            shift += 1
        return left << shift
    
solution = Solution()
solution.rangeBitwiseAnd(26, 30)
    

````

371. Sum of Two Integers

````python
''' 
This is the classic **â€œSum of Two Integers without using + or -â€** problem.
Letâ€™s go **from intuition â†’ bit rules â†’ code â†’ dry run**, very slowly.

---

## ðŸ”¢ Problem

```python
Input: a = 1, b = 2
Output: 3
```

You are **not allowed** to use `+` or `-`.

---

## ðŸ§  Core Idea (how addition works internally)

When you add two numbers in binary, **two things happen**:

### 1ï¸âƒ£ Sum without carry â†’ XOR

```text
1 ^ 0 = 1
1 ^ 1 = 0   (carry generated)
```

So:

```
a ^ b  â†’ partial sum (no carry)
```

---

### 2ï¸âƒ£ Carry calculation â†’ AND + shift

```text
1 & 1 = 1   (carry)
```

Carry must be added **one bit left**:

```
(a & b) << 1
```

---

## ðŸ” Repeat until no carry remains

* Add **partial sum** and **carry**
* Stop when carry becomes `0`

---

## âœ… Correct Implementation (Python-safe)

Python integers are unlimited length, so we must **simulate 32-bit signed integers**.

```python
class Solution:
    def getSum(self, a: int, b: int) -> int:
        MASK = 0xFFFFFFFF      # 32-bit mask
        MAX_INT = 0x7FFFFFFF   # max positive 32-bit int

        while b != 0:
            carry = (a & b) & MASK
            a = (a ^ b) & MASK
            b = (carry << 1) & MASK

        # handle negative numbers
        return a if a <= MAX_INT else ~(a ^ MASK)
```

---

## ðŸ§ª Dry Run (a = 1, b = 2)

Binary:

```
a = 001
b = 010
```

### Iteration 1

```
sum   = a ^ b = 011 (3)
carry = a & b = 000
```

Carry = 0 â†’ stop

âœ” Answer = `3`

---

## ðŸ§ª Another example (with carry)

### a = 3, b = 5

```
3 = 011
5 = 101
```

### Iteration 1

```
sum   = 110 (6)
carry = 001 << 1 = 010
```

### Iteration 2

```
sum   = 110 ^ 010 = 100 (4)
carry = 110 & 010 = 010 << 1 = 100
```

### Iteration 3

```
sum   = 100 ^ 100 = 000
carry = 100 & 100 = 100 << 1 = 1000
```

Eventually:

```
result = 8
```

âœ” Correct: `3 + 5 = 8`

---

## ðŸ”‘ Why this always works

| Operation    | Purpose                         |
| ------------ | ------------------------------- |
| XOR (`^`)    | Adds bits without carry         |
| AND (`&`)    | Finds carry bits                |
| SHIFT (`<<`) | Moves carry to correct position |

---

## ðŸ§  One-line explanation (interview-ready)

> â€œWe simulate binary addition using XOR for sum and AND + shift for carry, repeating until no carry remains.â€

---

## â± Complexity

* **Time:** `O(1)` (max 32 iterations)
* **Space:** `O(1)`

---

## ðŸŽ¯ Final takeaway

* This mimics **how CPUs add numbers**
* Very popular **bit-manipulation interview problem**
* Once understood, itâ€™s unforgettable

If you want next:

* Simpler version assuming only positive numbers
* Visual bit animation
* Why Python needs masking but Java/C++ donâ€™t
* Similar problems using XOR

Just tell me ðŸ‘

'''
````

````python
''' 
Input: a = 1, b = 2
Output: 3
'''

class Solution:
    def getSum(self, a: int, b: int) -> int:
        MASK = 0xFFFFFFFF
        MAX_INT = 0x7FFFFFFF

        while b != 0:
            carry = (a & b) & MASK
            a = (a ^ b) & MASK 
            b = (carry << 1) & MASK
        return a if a <= MAX_INT else ~(a ^ MASK)

Solution().getSum(1, -2)
````

260. Single Number III

````python
''' 
Input: nums = [1,2,1,3,2,5]
Output: [3,5]
Explanation:  [5, 3] is also a valid answer.
'''
from typing import List

class Solution:
    def singleNumber(self, nums: List[int]) -> List[int]:
        xor = 0 
        for n in nums:
            xor = xor ^ n 
        
        diff_bit = 1 
        while not (diff_bit & xor):
            diff_bit = diff_bit << 1

        a, b = 0, 0 

        for num in nums:
            if num & diff_bit:
                a = a ^ num 
            else:
                b = b ^ num 
        return [a, b]
    
solution = Solution()
solution.singleNumber([1,2,1,3,2,5])
````
